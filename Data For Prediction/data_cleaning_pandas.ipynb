```python
def concat_above_999(x):
    if pd.isna(x):  # Properly check for NaN values
        return None
    parts = str(x).split(',')
    return ''.join(parts) if len(parts) > 1 else str(x)  # Handle cases without a comma

concat_above_999("1,778 Sq. Yd.")
```

```python
df['area'] = df['area'].apply(concat_above_999)
```

```python
def convert(x):
    if str(x).split(" ")[-1] == "Marla":
        return (float(str(x).split(" ")[0]) * 272.25)
    if str(x).split(" ")[-1] == "Kanal":
        return (float(str(x).split(" ")[0])*5445)
    if str(x).split(" ")[-1] == "Yd.":
        return (float(str(x).split(" ")[0])*9)
        
# convert("1 Marla")
# convert("1 Kanal")
convert("1 Yd.")
```

```python
def convert_prices_unit_to_pkr(x):
    if str(x).split(" ")[-1]=="Crore":
        return float(str(x).split(" ")[0])*10000000
    if str(x).split(" ")[-1]=="Lakh":
        return float(str(x).split(" ")[0])*100000
    if str(x).split(" ")[-1]=="Arab":
        return float(str(x).split(" ")[0])*1000000000
    return None

convert_prices_unit_to_pkr("15 Lakh")
    
```

```python
def filter_on_threshold(x):
    if (x['location_city'] == "Karachi") and ((x['Square_per_ft'] > 4000) and (x['Square_per_ft'] < 35000)):
        return True
    elif (x['location_city'] == "Islamabad") and ((x['Square_per_ft'] > 9000) and (x['Square_per_ft'] < 60000)):
        return True
    elif (x['location_city'] == "Rawalpindi") and ((x['Square_per_ft'] > 2500) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Lahore") and ((x['Square_per_ft'] > 4000) and (x['Square_per_ft'] < 24000)):
        return True
    elif (x['location_city'] == "Multan") and ((x['Square_per_ft'] > 2000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Gujranwala") and ((x['Square_per_ft'] > 2000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Faisalabad") and ((x['Square_per_ft'] > 2000) and (x['Square_per_ft'] < 20000)):
        return True  
    elif (x['location_city'] == "Peshawar") and ((x['Square_per_ft'] > 2800) and (x['Square_per_ft'] < 20000)):
        return True  
    elif (x['location_city'] == "Hyderabad") and ((x['Square_per_ft'] > 2500) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Gwadar") and ((x['Square_per_ft'] > 1500) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Wah") and ((x['Square_per_ft'] > 2000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Sialkot") and ((x['Square_per_ft'] > 3000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Bahawalpur") and ((x['Square_per_ft'] > 1500) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Taxila") and ((x['Square_per_ft'] > 3000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Punjab"):
        return True
    elif (x['location_city'] == "Jhelum") and ((x['Square_per_ft'] > 2000) and (x['Square_per_ft'] < 20000)):
        return True
    elif (x['location_city'] == "Abbottabad") and ((x['Square_per_ft'] > 2500) and (x['Square_per_ft'] < 20000)):
        return True
    else:
        return False
```

```python
df['Filter_by_area/price'] = df.apply(filter_on_threshold, axis=1)
```
